```python
dfp.describe()
```

```python
sns.jointplot(x = dfp['budget'], y = dfp['production_year'], data = dfp)
plt.savefig('yearvsbudget.png')
```

Firstly, this jointplot shows the differences in budgets from year to year. As we can budgets have relatively stayed the same from 2006 to 2017.

```python
dfcopy = dfp.copy()
df2017 = dfcopy[dfcopy['production_year'] > 2016]
df20172 = df2017[df2017['budget'] > 100000000]
sns.set(font_scale = 0.5)
sns.barplot(x = df20172['budget'], y = df20172['genres'], color='orange',errorbar = None)
plt.title("Genre Vs Budget 2017")
plt.savefig('genrevsbudget2017.png',bbox_inches="tight",dpi=200)
```

```python
dfcopy2 = dfp.copy()
df2006 = dfcopy2[dfcopy2['production_year'] < 2007]
df20062 = df2006[df2006['budget'] > 100000000]

sns.set(font_scale = 0.5)
sns.barplot(x = df20062['budget'], y = df20062['genres'],color='orange', errorbar = None)
plt.title("Genre Vs Budget 2006")
plt.savefig('genrevsbudget2006.png', bbox_inches="tight",dpi=200)
```

Two barplots showing the top budgets for films in 2006 and 2017(budgets over 150 million), we can see that action is the dominate genre in terms of budgets amongst films. There are some outliers but generally it is Action that takes the highest budget. We can see the subgenres are different but there main genre is still Action. This I believe to be due to Action films requiring stunts, extra staff in the form of stunt double, props used like cars etc. These films require alot of planning and upfront cost. 

```python
dfcopy = dfp.copy()
df2017 = dfcopy[dfcopy['budget'] > 200000000]
sns.set(font_scale = 0.5)
sns.jointplot(x = df2017['directors'], y = df2017['budget'])
plt.xticks(rotation=90)
plt.savefig('directorsbudget.png')
```

This is a jointplot of what directors had the highest budgets between 2006 and 2017.

Conclusion
As we can see the data shows that we over the the time period of 2006 to 2017 movie budgets relatively stay the same from year to year. Looking at the two barplots we find that Action movies almost always yeild the highest budgets amongst its counterparts and that begs the question do genres impact a films budget? looking at the data we can see that a genre can impact a films budget immensely, for example an action move will require a higher budget due to stunts, explosions etc compared to a drama film which is purely acting.
