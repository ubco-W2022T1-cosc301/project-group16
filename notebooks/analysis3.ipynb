Firstly, this jointplot shows the differences in budgets from year to year. As we can budgets have relatively stayed the same from 2006 to 2017.

```python
dfcopy = dfp.copy()
df2017 = dfcopy[dfcopy['production_year'] > 2016]
sns.set(font_scale = 0.5)
sns.barplot(x = df2017['budget'], y = df2017['genres'])
```

Two Bar plots thatlook at the genre of top 100 movies based on there budgets from the year 2006 and 2017. We can see a correlation with the highest budgets in both going to the action/adventure/sci-fi category. We see the budget for fantasy movies have dropped since 2006. This information provides a look at the how movie budget can be changed as societies perfernces change as well ie: fantasy movies becoming less popular leading to lower budget while action movies becoming increasingly popular so therefore budget of these films increases.

```python
dfcopy = dfp.copy()
df2017 = dfcopy[dfcopy['budget'] > 200000000]
sns.set(font_scale = 0.5)
sns.displot(x = df2017['directors'], y = df2017['genres'], hue=df2017['budget'])
plt.xticks(rotation=90)
```

This is a jointplot of what directors had the highest budgets between 2006 and 2017 and what genre they directed.

Conclusion
As we can see the data shows that we over the the time period of 2006 to 2017 movie budgets relatively stay the same from year to year. The data that changed the most was the genre budgets, from 2006 to 2017 we seen the budget of certain genres decline and others rise. Do genres impact a films budget? looking at the data we can see that a genre can impact a films budget immensely, for example an action move will require a higher budget due to stunts, explosions etc compared to a drama file which is purely acting.
